#### Step 3: Customize the User Prompt

Find the best career suited to you

````python
system_message = "You are a career strategist with a positive but grounded approach & energy when working with someone. It should feel like a personalized coach who understood my background and was committed to helping me with my job search."

user_prompt = """
Help me find my ideal career based on the following:

1. My Skillset
- Curiosity-honed prowess in data related technologies (ingestion, storage, etl, analysis, visualization) 
- Willingness to listen closely to others 
- Enthusiasm for solving problems all the way through 

2. My current industry
- Independent data and AI projects

3. My interests
- Narratives, typically consumed in the form of public radio, podcasts, novels, anime (recently)
- Health and fitness, for myself and those around me
- Trying to get better at things I'm not proficient at yet, like drawing and music and reaching out

4. My objectives
- Find current job openings that could be a mutually good fit
- Craft job applications for each position that's a good fit & submit them in <30 min
- Plan a career trajectory that will lead in the direction of ontology & graph-based data analysis and data governance

Now based on this, give me:
- A list of job titles I may want to search for NOW (as in, use trends relevant to the last 3-6 months)
- A list of industries and sub-industries that I may want to keep a close eye on PLUS advise on how to do that
- A list of PICK-ME-UPS that will give me energy to do what I need to do in those moments where I'm low energy and/or bored
"""

````

````python
client = OpenAI()

response = client.chat.completions.create(
    model = "chatgpt-4o-latest",
    messages = [{"role": "system", "content": system_message},{"role": "user", "content": user_prompt}],
    temperature=0.8
)
````

````python
print(response.choices[0].message.content)

````

````python
# Step 1: I used the whole response to my original question & tell the 
career_suggestion = response.choices[0].message.content # My whole response (career suggestions, industries, motivational tips) from above

# Step 2: Write a reasoning prompt that will help the model plan how to pursue this career
# Your prompt must:
# - Refer to the user's profile + their chosen career
# - Ask for a step-by-step plan
# - Include at least 3 types of guidance (e.g. skills, resources, milestones)
reasoning_prompt = f"""
    Take the '**Suggested Job Titles**' and '**High Potential Industries** for me' from the text within '***' delimiters below. Your goal is to 
    Specifically consider the person's skillset and desired career and produce a 3 Week Implementation Plan to guide them, including daily and weekly milestones. 
    That detailed 3 Week Implementation plan should also include Step-by-Step guidance that builds upon the person's strengths and suggests other skills 
    or 'gap areas' they should augment and list resources to reach their milestones.

    ***
    {career_suggestion}
    ***
"""

completion = client.chat.completions.create(
    model="gpt-5-mini",
    messages=[
        {"role": "system", "content": system_message},
        {"role": "user", "content": reasoning_prompt}
    ]
) 

````
