## Week 5, Day 7 - Agentic Loops 101

Let's make a version of the Dice Roll capabilities that's basic but "agentic" ... 

```python
import random
from openai import OpenAI
 
 
client = OpenAI()

def roll_dice():
    return random.randint(1,6)

target = 6
max_iterations = 1

history = ""
iteration = 0

while iteration <= max_iterations:
    iteration +=1

    instructions = f"""
    You are playing a dice game. You need to roll a {target} on the dice.

    You've made the following rolls so far:
    {history if history else "No rolls yet."}

    Based on what you have rolled so far, decide your next action, either Option A or B:
    
    - Option A: If you have achieved the target with the latest roll, respond with exactly: DONE: <message describing how you reached the target>

    - Option B: If you have not yet achieved the target, respond with exactly: ROLL THE DICE
    """

    response = client.chat.completions.create(
        model = 'gpt-4.1-mini',
        messages=[
            {"role": "user", "content": instructions}
        ]
    )

    result = response.choices[0].message.content 


    if "DONE" in result:
        print(result)
        break
    
    if "ROLL THE DICE" in result:
        roll = roll_dice()
        print(f"🎲 roll was: {roll}\n")
        history = f"Roll {iteration}: {roll}\n"

print(f"Game ended after {iteration-1} iterations.")
```
