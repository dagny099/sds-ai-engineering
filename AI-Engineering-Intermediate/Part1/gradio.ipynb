```python
## Step 1: Import Libraries and API Keys

import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_API_KEY is None:
    raise Exception("API key is missing")


```

## Step 2: Basic UI Interface

```python
import gradio as gr
```

```python
def respond_basic(message, history):
    return f"Me encanta la IA \n **msg: {message}, **history: {history}"
```

```python
gr.ChatInterface(fn=respond_basic).launch(inbrowser=True)
```

## Step 3: Add OpenAI Functionality

```python
import openai

def respond_ai(message, history):
    msgs = [{"role": "assistant", "content": "You are a helpful assistant!"}] + history + [{"role": "user", "content": message}]
    client = OpenAI(api_key=OPENAI_API_KEY)
    response = client.chat.completions.create(
                model = "gpt-4.1-mini",
                messages = msgs
            )
    reply = response.choices[0].message.content
    return reply

gr.ChatInterface(fn=respond_ai).launch(inbrowser=True)
```

```python

history = [
    {"role": "assistant", "content": "I am happy to provide you that report and plot."},
]

with gr.Blocks() as demo:
    gr.Chatbot(history)

demo.launch()
```
