```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_API_KEY is None:
    raise Exception("Key is not set")

client = OpenAI(api_key=OPENAI_API_KEY)

```

```python
from litellm import completion

```

```python
messages = [{"role": "user", "content": "Where is the Amazon River located?"}]

response = completion(
    model = "openai/gpt-5-nano",
    messages=messages
)
```

```python
display(Markdown(response.choices[0].message.content))
```

### Step 3: LLM Call Info


```python
import json
```

```python
print(json.dumps(response.model_dump(), indent=2))
```

```python
print(f"prompt tokens: {response.usage.prompt_tokens}")
print(f"prompt tokens: {response.usage.completion_tokens}")
print(f"prompt tokens: {response.usage.total_tokens}")
```

```python
print(json.dumps(response._hidden_params, indent=2))
```

```python
print(f"cost: ${response._hidden_params['response_cost']:.6f}")
```
